```python
import numpy as np
def Find_Dist(a, b):
    # convert data to numpy for easy element-wise operation
    a = np.array(a)
    b = np.array(b)

    # Return Euclidean distance( think pythagoras)
    dist = (a - b)**2
    x = np.sum(dist)
    x = np.sqrt(x)
    return x
```

```python
Find_Dist([7, 5], [3, 2])
```

```python
def Majority_count(array):
    ''' returns mode'''
    counts = {}
    # Counts frequency of each data in array
    for x in array:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    most_freq = max(counts.values())
    # loop through counts if counts of key equals most_freq, return key
    for data, count in counts.items():
        if count == most_freq:
            return data
```

```python
Majority_count([1, 5, 3, 1, 2, 2, 3, 5, 1])
```

```python
def knn_Classified(points, outcome, newpoint):
# Start out with zero array, we will later fill with actual distance
    dists = np.zeros(len(outcome)) 
    # compute distance between each point and new point
    for idx in range(len(points)):
        dists[idx] =  Find_Dist(points.iloc[idx], newpoint)
        # dists[idx] =  Find_Dist(points[idx], newpoint)

    # Sort distances based on index
    sorted_dist = np.argsort(dists)

    #solve for k
    k = int(np.sqrt(len(dists)))
    # Requied data for prediction
    required = outcome[sorted_dist[0 : k]]

    # Return class with majority
    return Majority_count(required)
```
